# 01-segment-idox.ipynb
This notebook is for segmenting the idox crystals from each other and from the binder material in the IDOX images obtained by Yeager et al. at Los Alamos National Lab (LANL).

# Import packages

```python
# Standard library imports
from pathlib import Path
# Third-party imports
import imagecodecs  # dependency required for loading compressed tif images
import imageio as iio
import matplotlib.pyplot as plt
import numpy as np
from skimage import util
```

# Load data

```python
project_dir = Path('..').resolve()
print(project_dir)
```

```python
data_dir = Path(project_dir / 'data/MOCK-IDOX_33')   
img_paths = [path for path in data_dir.glob('*.tif')]
img_paths.sort()
print(f'{len(img_paths)=}')
print(f'{img_paths[500]=}')
```

```python
img_500 = iio.imread(img_paths[500])
img_500 = util.img_as_float(img_500)
print(f'{img_500.shape=}')
fig, ax = plt.subplots(figsize = (8, 8))
ax.imshow(img_500)
ax.axis('off')
plt.show()
```
